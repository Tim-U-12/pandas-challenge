```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

# Local Government Area Summary

```python
# LGA Summary
total_unique_schools = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["Student ID"].count()
total_budget = school_data_complete["budget"].unique().sum()
avg_math_score = school_data_complete["maths_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()
pass_math = school_data_complete.loc[school_data_complete["maths_score"] >= 50,"maths_score"].count() / total_students * 100
pass_english = school_data_complete.loc[school_data_complete["reading_score"] >= 50,"reading_score"].count() \
                / total_students * 100
pass_overall = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50),'Student ID'].count()\
                / total_students * 100

total_unique_schools
```

```python
# LGA Summary Table
LGA_summary = pd.DataFrame({"Total Schools":[total_unique_schools],
                           "Total Students":[total_students],
                           "Total Budget": [total_budget],
                           "Average Maths Score": avg_math_score,
                           "Average Reading Score":avg_reading_score,
                           "% Passing Maths": pass_math,
                           "% Passing Reading": pass_english,
                           "% Overall Passing": pass_overall})
LGA_summary
```

# School Summary

```python
def check_unique(alist: list) -> int:
    if len(alist) == 1:
        return alist[0]
    else:
        raise Exception("Error")
```

```python
groupby_school = school_data_complete.groupby(['school_name'])
school_type = groupby_school["type"].unique()
school_type = school_type.apply(check_unique)

total_students = groupby_school.size()
total_school_budget = groupby_school["budget"].unique()
total_school_budget = total_school_budget.apply(check_unique)

per_student_budget = total_school_budget / total_students
school_avg_math_score = groupby_school["maths_score"].mean()
school_avg_reading_score = groupby_school["reading_score"].mean()

school_pass_math = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(['school_name']).count()
school_pass_math = school_pass_math['maths_score'] / total_students * 100
school_pass_read = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(['school_name']).count()
school_pass_read = school_pass_read["reading_score"] / total_students * 100 

school_pass_overall = school_data_complete[(school_data_complete["reading_score"] >= 50) & (school_data_complete["maths_score"] >= 50)].groupby(['school_name']).count()
school_pass_overall = school_pass_overall["Student ID"] / total_students * 100
```

```python
school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_students, 
    "Total School Budget": total_school_budget,
    "Per Student Budget": per_student_budget,
    "Average Maths Score": school_avg_math_score,
    "Average Reading Score": school_avg_reading_score,
    "% Passing Maths" : school_pass_math,
    "% Passing Reading": school_pass_read,
    "% Overall Passing": school_pass_overall
})
school_summary
```

# Highest-Performing Schools (by % Overall Passing)

```python
top_schools = school_summary.sort_values(by="% Overall Passing", axis=0, ascending=False)[0:5]
top_schools
```

# Lowest-Performing Schools (by % Overall Passing)

```python
bottom_schools = school_summary.sort_values(by="% Overall Passing", axis=0, ascending=False)[-5:]
bottom_schools
```

# Maths Scores by Year

```python
year_nine_math_score = school_data_complete.loc[school_data_complete["year"] == 9]
year_nine_math_score = year_nine_math_score.groupby(["school_name"]).mean()

year_ten_math_score = school_data_complete.loc[school_data_complete["year"] == 10]
year_ten_math_score = year_ten_math_score.groupby(["school_name"]).mean()

year_eleven_math_score = school_data_complete.loc[school_data_complete["year"] == 11]
year_eleven_math_score = year_eleven_math_score.groupby(["school_name"]).mean()

year_twelve_math_score = school_data_complete.loc[school_data_complete["year"] == 12]
year_twelve_math_score = year_twelve_math_score.groupby(["school_name"]).mean()

math_score_by_year = pd.DataFrame({
    "Year 9":year_nine_math_score["maths_score"],
    "Year 10":year_ten_math_score["maths_score"],
    "Year 11":year_eleven_math_score["maths_score"],
    "Year 12":year_twelve_math_score["maths_score"]
})

math_score_by_year
```

# Reading Score by Year

```python
reading_score_by_year = pd.DataFrame({
    "Year 9":year_nine_math_score["reading_score"],
    "Year 10":year_ten_math_score["reading_score"],
    "Year 11":year_eleven_math_score["reading_score"],
    "Year 12":year_twelve_math_score["reading_score"]
})

reading_score_by_year
```

# Scores by School Spending

```python
bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Per Student Budget"], bins, labels=group_names, include_lowest=True)
groupby_spending_ranges = school_summary.groupby("Spending Ranges (Per Student)").mean()

del groupby_spending_ranges["Total Students"]
del groupby_spending_ranges["Total School Budget"]
del groupby_spending_ranges["Per Student Budget"]

groupby_spending_ranges.round(2)
```

# Scores by School Size

```python
bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary["School Size"] = pd.cut(school_summary["Total Students"], bins, labels=group_names, include_lowest=True)
groupby_school_size = school_summary.groupby("School Size").mean()

del groupby_school_size["Total Students"]
del groupby_school_size["Total School Budget"]
del groupby_school_size["Per Student Budget"]

groupby_school_size
```
